```python
import numpy as np
data = np.loadtxt("train.csv",delimiter =",")
```

```python
x=data[:,0:9]
y=data[:,10]
x
```

```python
x=data[:,0:9]
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
```

```python
alg1.fit(x,y)
```

```python
m = alg1.coef_[0]
c = alg1.intercept_
```

```python
data2 = np.loadtxt("test.csv",delimiter=",")
x_test = data2[:,0:9]
```

```python
y_pred = alg1.predict(x_test)
y_pred
```

```python
def predict(x,m,c):
    return m*x + c
```

```python
y_pred_2 = predict(x_test,m,c)
y_pred_2
```
